### **Rôle principal du routeur de Prescriptions :**
Ce fichier définit les points d'entrée (endpoints) de votre API concernant les ordonnances. Il sert de pont entre l'interface utilisateur (front-end) et la logique métier contenue dans prescription_service.py. Il gère la validation des requêtes HTTP, le contrôle d'accès et le formatage des réponses.

### **Étapes et routes implémentées :**

##### **Étape 1 : Création et déclenchement de l'analyse (POST /)**

- Cette route permet d'enregistrer une nouvelle ordonnance et de déclencher immédiatement le moteur de règles cliniques (CDS).

- Contrôle d'accès : Elle vérifie strictement que l'utilisateur connecté possède le rôle de médecin (Role.DOCTOR) ou de pharmacien (Role.PHARMACIST) avant d'autoriser l'action.

- Formatage : Après la création, le script "aplatit" la structure complexe des alertes (qui sont rattachées à chaque ligne de prescription) pour les regrouper dans un objet unique CdsResponse, facilitant ainsi leur affichage côté front-end.

##### **Étape 2 : Consultation détaillée (GET /{prescription_id})**

- Ce point de terminaison permet de récupérer une ordonnance existante avec l'ensemble de ses lignes et de ses alertes associées.

- Il inclut une gestion d'erreur standard : si l'identifiant n'existe pas en base, il renvoie une erreur HTTP 404.

- La réponse est structurée avec le statut global de l'ordonnance (ex: "safe" ou "alerts") et le décompte total des avertissements.

##### **Étape 3 : Historique par patient (GET /patient/{patient_id})**

- Cet endpoint liste toutes les prescriptions associées à un patient spécifique.

- Il intègre un mécanisme de pagination (skip et limit, fixée à 20 par défaut) afin d'optimiser les performances lors du chargement des dossiers patients volumineux.

##### **Étape 4 : Suivi personnel du praticien (GET /doctor/mine)**

- Cette route est conçue pour le tableau de bord du médecin, lui permettant de consulter l'historique des prescriptions qu'il a lui-même générées.

- La requête SQL sous-jacente précharge efficacement les données liées (lignes de prescription et alertes) via selectinload pour éviter le problème des requêtes N+1.

- Les résultats sont triés chronologiquement par date de création décroissante (les plus récentes en haut) et la pagination est plafonnée à un maximum de 100 éléments par requête.
