Ce document présente un projet de système intelligent de sécurité médicamenteuse basé sur l'IA pour réduire les erreurs et l'alerte fatigue dans le contexte médical marocain.

<br>

**1. Contexte général du projet SafeRx AI :**

Ce projet vise à développer un assistant intelligent pour la sécurité médicamenteuse, intégrant une architecture RAG optimisée pour réduire les erreurs et alertes inutiles.

- Système basé sur une architecture modulaire avec plusieurs couches technologiques.
- Objectif principal : sécuriser le circuit du médicament au Maroc.
- Inclut des interfaces pour médecins, pharmaciens et administrateurs.
- Utilise des technologies modernes comme React, FastAPI, LangChain, Ollama, et Scikit-learn.
- Infrastructure déployée via Docker, Airflow, Prometheus, et Grafana.

<br>

**2. Problématique et enjeux liés à la sécurité médicamenteuse :**

Les erreurs médicamenteuses représentent un enjeu critique, avec des risques graves pour la sécurité des patients.

- 1 patient sur 30 subit des effets indésirables liés aux médicaments (OMS).​
- Plus d’un quart de ces cas sont graves ou vitaux.
- 25% des hospitalisations sont dues à des erreurs médicamenteuses.
- Absence de systèmes numériques de vérification au Maroc.
- Jusqu’à 70% des alertes répétitives ou non pertinentes sont ignorées par les médecins.​

<br>

**3. Objectifs et solutions du projet SafeRx AI :**

Ce projet vise à améliorer la sécurité clinique, l’explicabilité de l’IA, et à réduire l’alerte fatigue.

- Détection automatique des interactions, contre-indications, et redondances de DCI.
- Enrichissement des alertes avec des justifications textuelles via LLM.
- Réduction du bruit d’alertes grâce à un modèle ML priorisant celles cliniquement significatives.​
- Enregistrement exhaustif de chaque décision pour la traçabilité et l’audit.​
- Mise à jour automatisée des données via Apache Airflow.
- Surveillance proactive avec Prometheus et Grafana.

<br>

**4. Architecture et composants techniques :**

Une architecture modulaire avec plusieurs couches technologiques.

- Interfaces utilisateur : React, Tailwind CSS pour dashboards médecins, pharmaciens, admin.
- API Gateway : FastAPI avec JWT/OAuth2, SQLAlchemy.
- Services métier : moteur CDS, IA RAG, ML, audit.
- Couche IA : LangChain, Ollama (Llama3.1), Gemini fallback, Scikit-learn.
- Bases de données : PostgreSQL pour médicaments, interactions, audit.​
- Infrastructure : Docker Compose, Airflow, Prometheus, Grafana, CI/CD via GitHub Actions.

<br>

**5. Base de données médicamenteuse et sources :**

Le socle de connaissances cliniques est alimenté par scraping de medicament.ma et le Thésaurus ANSM 2023.

- 6249 composants DCI, 5006 médicaments, 3168 interactions.
- Extraction automatisée et mise à jour hebdomadaire.
- Données cryptées au repos avec AES-256.

<br>

**6. Moteur CDS et règles cliniques :**

Un moteur basé sur 7 règles cliniques pour détecter les risques majeurs.

- Détection d’allergies, redondances, contre-indications, insuffisance rénale.
- Règles spécifiques pour allergies majeures, posologie, interactions, etc.
- Règles mineures pour alertes informatives comme effets psychoactifs.

<br>

**7. Pipeline RAG pour explication des alertes :**

Génération d’explications transparentes via RAG.

- Extraction du contexte patient et médicaments.​
- Construction de prompts via templates.
- Inférence avec Ollama ou Gemini.
- Filtrage qualité et stockage des explications.
- Rejet si réponse trop courte ou trop longue.

<br>

**8. Validation et traçabilité des décisions :**

Analyse sémantique pour valider les justifications d’override.

- Justification par le médecin en texte libre.
- Classification en "Valide" ou "Nuisible".
- Stockage dans l’audit trail pour traçabilité.​
- Boucle d’apprentissage pour améliorer le système.

<br>

**9. Modèle ML anti-alert fatigue :**

Réduction de la surcharge cognitive des médecins.

- Utilise une régression logistique pour prédire l’ignorance d’une alerte.​
- Facteurs : type d’alerte, heure, niveau de sévérité, jour.
- Score d’IA : 0.82 indique forte probabilité d’ignorance.
- Objectif : prioriser les alertes pertinentes et réduire la fatigue.

<br>

**10. Audit trail et supervision :**

Traçabilité complète pour responsabilité et amélioration continue.​

- Enregistrement immuable de chaque décision.
- Données cryptées, latence < 10 ms.
- Boucle d’apprentissage basée sur les données d’audit.
- Table d’audit : décision, contexte, justification, validation IA, métadonnées.

<br>

**11. Monitoring et supervision en temps réel :**

Suivi de la performance via Prometheus et Grafana.

- Indicateurs clés : alertes CDS, prescriptions, enrichissement RAG, décisions d’audit.
- Surveillance de la latence API et ressources système.
- Collecte granulaire avec cAdvisor pour optimisation proactive.

<br>

**12. Pipeline CI/CD et déploiement :**

Automatisation du cycle de développement et déploiement.

- Analyse statique, tests unitaires avec Pytest, Mock.
- Docker pour build et déploiement.
- Déclenchement automatique via GitHub Actions.
- Processus intégré pour garantir la qualité et la rapidité des mises en production.
